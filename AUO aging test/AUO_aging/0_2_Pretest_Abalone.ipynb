```python
rabalone = pyreadr.read_r('C:/Users/Darui Yen/OneDrive/桌面/0629_aging/abalone.rda') 
rabalone = rabalone['abalone']

strip_class = []
for i in range(len(rabalone)):
    strip_class.append(rabalone.Class.values[i].strip())
rabalone['Class'] = strip_class

rabalone = rabalone.replace(to_replace = ['negative', 'positive'], value = [0, 1])

rabalone.head()
```

```python
# data from UCI machine learning repository
abalone = pd.read_csv('C:/Users/Darui Yen/OneDrive/桌面/0629_aging/abalone.data', sep = ',', header = None)
abalone.columns = ['sex', 'length', 'diameter', 'height', 'whole weight', 'shucked weight', 'viscera weight', 'shell weight', \
                  'rings']

#classification by rings level
abalone = pd.get_dummies(abalone, columns = ['sex'])

data_num = abalone.shape[0]
cate1 = []
cate2 = []
for i in range(data_num):
    if abalone['rings'][i] < 7:
        cate1.append('low')
        cate2.append('young')
    elif abalone['rings'][i] < 9:
        cate1.append("low")
        cate2.append('old')
    elif abalone['rings'][i] > 10:
        cate1.append("high")
        cate2.append('old')
    else:
        cate1.append("medium")
        cate2.append('old')

abalone['levels'] = cate1
abalone['age'] = cate2
abalone = abalone.replace(to_replace = ['old', 'young'], value = [0, 1])

abalone.head()
```

### Imbalance ratio

```python
major1 = sum(x == 0 for x in rabalone['Class'])
minor1 = sum(x == 1 for x in rabalone['Class'])
imratio1 = major1/minor1
print('rabalone:\n', 'Imbalance Ratio:', imratio1, '\n')

major2 = sum(x == 0 for x in abalone['age'])
minor2 = sum(x == 1 for x in abalone['age'])
imratio2 = major2/minor2
print('abalone:\n', abalone.groupby('rings').count().iloc[:,0])
print('\n', abalone.groupby ('levels').count().iloc[:,0])
print('\n', abalone.groupby('age').count().iloc[:,0], '\nImbalance Ratio:', imratio2)
```
