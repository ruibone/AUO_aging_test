```python
###real data for aging test###
data = {}
for i in range(1, 13):
    data[f"set{i}"] = pd.read_csv(f"C:/Users/Darui Yen/OneDrive/桌面/Regressor_data/DS_reg_{i}.csv")
    data[f"set{i}"]["F1"] \
    = (2*data[f"set{i}"]["Recall"]*data[f"set{i}"]["Precision"]) / (data[f"set{i}"]["Recall"]+data[f"set{i}"]["Precision"])
    
###simulated data for contour plot###
x = np.linspace(0.01, 1, 100)
y = np.linspace(0.01, 1, 100)
X, Y = np.meshgrid(x, y)
```

```python
data["set10"].head()
```

```python
def score1(r, ag):
    eff = r/ag
    eff_one = eff > 1
    score = r*np.log(1+eff)
    score = score*eff_one
    return score


def score2(r, ag):
    eff = r/ag
    eff_one = eff > 1
    score = (r/(1.5-r))*np.log(1+eff)
    score = score*(score > 0)
    score = score*eff_one
    return score


def score3(r, ag):
    eff = r/ag
    eff_one = eff > 1
    score = (r-0.5)*np.log(2+eff)
    score = score * (score > 0)
    score = score*eff_one
    return score


def contourplot(x,y,z):
    fig = plt.figure()
    plt.contourf(x, y, z, levels = 50, cmap = "gist_rainbow", alpha = 0.8)
    plt.colorbar()
    plt.plot([0, 1], [0, 1], 'k-', lw = 2)
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.xlabel("Recall")
    plt.ylabel("Aging Rate")
    fig.set_size_inches(9, 5)
    
    return fig


def findthreshold(set_num, data = data, threshold = np.linspace(1, 0.05, 20), target = "class_pd"):
    thres_list = []
    index = 0
    if isinstance(threshold, float) == False: 
        for i in threshold:
            while not np.all([i >= data[f"set{set_num}"][target][index], data[f"set{set_num}"][target][index] > 0]) :
                index += 1
                if len(thres_list) > 1 and data[f"set{set_num}"][target][index] == 0:
                    break
            thres_list.append(index)
            index += 1

            if len(thres_list) < 1:
                index = 0
    else:
        while not np.all([threshold >= data[f"set{set_num}"][target][index], data[f"set{set_num}"][target][index] > 0]) :
            index += 1
        thres_list = [index]
        
    return thres_list


def findbestrecall(set_num, data = data, target = "Recall", mode = "train"): 
    index = 0
    while data[f"set{set_num}"][target][index] != 1:
        index += 1
    best_thres = data[f"set{set_num}"]["class_pd"][index]
    
    if  mode == "train":
        best = best_train = index
    else:
        best = best_test = findthreshold(set_num, threshold = best_thres, target = "class_pd.1")
    
    return best


def findbestF1(set_num, data = data, mode = "train"):
    best_train = np.nanargmax(data[f"set{set_num}"]["F1"])
    if mode == "train":
        best = best_train
    else:
        best_thres = data[f"set{set_num}"]["class_pd"][best_train]
        best = best_test = findthreshold(set_num, threshold = best_thres, target = "class_pd.1")
    
    return best


def lineplot(set_num, data = data, col_x = "Recall", col_y = "Precision", mode = "train"):
    
    if mode == "train":
        length = len(data[f"set{set_num}"]["Recall"]) - np.sum(np.isnan(data[f"set{set_num}"]["Recall"]))
        every005 = findthreshold(set_num)
        bestF1 = findbestF1(set_num)
        bestrecall = findbestrecall(set_num)
    else:
        length = len(data[f"set{set_num}"]["Recall.1"]) - np.sum(np.isnan(data[f"set{set_num}"]["Recall.1"]))
        every005 = findthreshold(set_num, target = "class_pd.1")
        bestF1 = findbestF1(set_num, mode = "test")
        bestrecall = findbestrecall(set_num, mode = "test")
        
    #median = math.floor(np.quantile(np.array(data[f"set{set_num}"]["Unnamed: 0"][1:length]), 0.5))
    
    plt.plot(data[f"set{set_num}"][col_x], data[f"set{set_num}"][col_y], "-")
    plt.xlabel(col_x)
    plt.ylabel(col_y)
    plt.title(f"DataSet {set_num}")
    
    plt.scatter(data[f"set{set_num}"][col_x][bestF1], data[f"set{set_num}"][col_y][bestF1], s = 120, c = "r")
    plt.scatter(data[f"set{set_num}"][col_x][bestrecall], data[f"set{set_num}"][col_y][bestrecall], s = 120, c = "g")
    
    for i in every005:
        if data[f"set{set_num}"]["F1"][i] >= np.nanmean(data[f"set{set_num}"]["F1"]):
            plt.scatter(data[f"set{set_num}"][col_x][i], data[f"set{set_num}"][col_y][i], s = 30, c = "r")
        else:
            plt.scatter(data[f"set{set_num}"][col_x][i], data[f"set{set_num}"][col_y][i], s = 30, c = "k")   
```

```python
###score1###
Z1 = score1(X, Y)
fig1 = contourplot(X, Y, Z1)
plt.title("Score 1")
fig1.savefig("score1.png", transparent = True)
```

```python
###### score2###
Z2 = score2(X, Y)
fig2 = contourplot(X, Y, Z2)
plt.title("Score 2")
fig2.savefig("score2.png", transparent = True)
```

```python
###score3###
Z3 = score3(X, Y)
fig3 = contourplot(X, Y, Z3)
plt.title("Score 3")
plt.plot([0.2,1], [0,0.8], "b-.", lw = 1)
fig3.savefig("score3.png", transparent = True)
```
