```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

### Website

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2019.html'
html = urlopen(url)
soup = BeautifulSoup(html)
```

### Per Game Stats

```python
header = [th.getText() for th in soup.findAll('table')[4].findAll('th', class_ = 'poptip center')]
table = soup.findAll('table')[4]
row = table.findAll('tr')[1:]
data = [[i.getText() for i in row[j].findAll('td')] for j in range(len(row))]
per_game = pd.DataFrame(data, columns = header)
per_game
```

### Advanced Stats

```python
table = soup.findAll('table')[7]
header = [i.getText() for i in table.findAll('tr')[0].findAll('th')]
row = table.findAll('tr')[1:]
data = [[i.getText() for i in row[j].findAll('td')] for j in range(len(row))]
advanced = pd.DataFrame(data, columns = header[1:])
advanced
```

### Salary

```python
url = 'https://hoopshype.com/salaries/2018-2019/'
html = urlopen(url)
soup = BeautifulSoup(html)

header = [th.getText() for th in soup.findAll('thead')[0].findAll('td')]
row = soup.findAll('tr')
data = [[td.getText()[10: -8] for td in row[i].findAll('td')] for i in range(1, len(row))]
salary = pd.DataFrame(data, columns = header).iloc[:, 1:]
salary.Team = salary.Team.str.strip()
salary
```
