```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
tomato = pd.read_csv("C:/Users/Darui Yen/OneDrive//桌面/crawler_pimpinellifolium_list.csv")
tomato
```

```python
m = Basemap(projection = 'merc', llcrnrlat = -25, urcrnrlat = 25, llcrnrlon = -110, urcrnrlon = -30, lat_ts = 20, 
            resolution = 'c')
m.shadedrelief()

m.drawparallels(np.arange(-90.,91.,30.))
m.drawmeridians(np.arange(-180.,181.,60.))
m.drawcountries()

plt.title("Germplasm Distribution")
num_accession = 0
for accession in range(tomato.shape[0]) :
    if tomato["longitude"][accession] > 0 or tomato["longitude"][accession] < 0 :
        m.scatter(x = tomato["longitude"][accession], y = tomato["latitude"][accession], latlon = True, c = "blue", s = 2)
        num_accession += 1
plt.show()
plt.savefig('C:/Users/Darui Yen/OneDrive//桌面/pim_geo.pdf')
```

```python
m = Basemap(projection = 'merc', llcrnrlat = -25, urcrnrlat = 25, llcrnrlon = -110, urcrnrlon = -30, lat_ts = 20, 
            resolution = 'c')
m.drawcoastlines()

m.drawparallels(np.arange(-90.,91.,30.))
m.drawmeridians(np.arange(-180.,181.,60.))

m.drawlsmask(ocean_color = "aqua")
m.drawmapboundary()
m.drawcountries()

plt.title("Germplasm Distribution")
num_accession = 0
for accession in range(tomato.shape[0]) :
    if tomato["longitude"][accession] > 0 or tomato["longitude"][accession] < 0 :
        m.scatter(x = tomato["longitude"][accession], y = tomato["latitude"][accession], latlon = True, c = "brown", s = 3)
        num_accession += 1
plt.show()
```
